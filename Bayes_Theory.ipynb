```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
#plot style
plt.style.use('fivethirtyeight')
# data for create dot grid
x_axis = np.linspace(0 ,1, 100)
y_axis = np.ones(100)
positive_observation = 20
real_positive_observation = 15
nevative_observation = 100-15
```

```python
fig, ax = plt.subplots(figsize=(18, 10))  

for point in x_axis:
    plt.scatter(x_axis, y_axis*point, s=5, c='black') 

for point in x_axis[:positive_observation]:
    plt.scatter(0, point, c='red', s=20, marker='s') 

for point in x_axis[:real_positive_observation]:
    plt.scatter(0, point, c='yellow', s=100, alpha=.4, marker='s') 

for point in x_axis[-5:]:
    plt.scatter(y_axis*point, x_axis, c='orange', s=80, alpha=.2, marker='s')

for point in x_axis[:nevative_observation]:
    plt.scatter(x_axis[-5], point, c='pink', s=80, alpha=.6, marker='s')

ax.set_title('Bayens Theory', fontsize=24, color='red')
ax.set_xlabel('Frequency', fontsize=18)
ax.set_ylabel('Propability of event', fontsize=18)

plt.show()
```
