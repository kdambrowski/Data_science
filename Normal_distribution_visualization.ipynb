```python
import numpy as np
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')

#data
Normal_distribution_scale_1 = np.random.normal(loc=0, scale= 1.0, size=100_000) 
Normal_distribution_scale_3  = np.random.normal(loc=0, scale= 3.0, size=100_000)
Normal_distribution_scale_10  = np.random.normal(loc=0, scale=10.0, size=100_000)
```

```python
plt.figure(figsize=(10, 7))
plt.hist(Normal_distribution_scale_10, bins=50, alpha=0.3, color='green', label= 'ND_SCALE10') 
plt.hist(Normal_distribution_scale_3, bins=50, alpha=0.3, color='blue', label= 'ND_SCALE3')
plt.hist(Normal_distribution_scale_1, bins=50, alpha=0.3, color='red', label= 'ND_SCALE1')
plt.legend()
plt.show()
```
