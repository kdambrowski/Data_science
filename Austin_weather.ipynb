```python
%matplotlib inline
import pandas as pd
import matplotlib.pyplot as plt
from pandas.plotting import register_matplotlib_converters
import numpy as np
register_matplotlib_converters()
import pylab
import scipy.stats as stats
import seaborn as sns
```

```python
url = 'https://raw.githubusercontent.com/matzim95/ML-datasets/master/austin_weather.csv'
austin_weather = pd.read_csv(url, parse_dates=True, index_col ='Date')
austin_weather.head()
```

```python
fig, ax = plt.subplots(2,1,sharex=True,figsize=(12,10))
n=50
ax[0].plot(austin_weather. index[:n], austin_weather['TempHighF'][:n].astype(float),
           marker = 'v', linestyle='--',color='r')
ax[0].plot(austin_weather. index[:n], austin_weather['TempAvgF'][:n].astype(float),
           marker = 'o',color='y')
ax[0].plot(austin_weather. index[:n], austin_weather['TempLowF'][:n].astype(float),
           marker = 'v', linestyle='dotted',color='b')

ax[1].plot(austin_weather. index[:n], austin_weather['DewPointHighF'][:n].astype(float),
           marker = 'v', linestyle='--',color='r')
ax[1].plot(austin_weather. index[:n], austin_weather['DewPointAvgF'][:n].astype(float),
           marker = 'o', linestyle='--',color='y')
ax[1].plot(austin_weather. index[:n], austin_weather['DewPointLowF'][:n].astype(float),
           marker = 'v', linestyle='dotted',color='b')

ax[0].set_title('Weather in Austin')

ax[1].set_xlabel('Date (year-month-day)')
ax[1].xaxis.set_tick_params(rotation=90)



ax[1].set_ylabel('Dew Point')
ax[0].set_ylabel('Temperature')
ax[0].annotate('Something has happend', xy=[pd.Timestamp('2014-01-06'),28],xytext=(pd.Timestamp('2014-01-09'), 22.5),arrowprops={'arrowstyle':'->', 'color':'green'})

plt.show()
```
