```python
network = models.Sequential()
network.add(Dense(512, activation='relu', input_shape=(28 * 28,)))
network.add(Dense(256, activation='tanh'))
network.add(Dense(10, activation='softmax'))
```

```python
network.compile(
    optimizer='adam',  # "rmsprop"
    loss='categorical_crossentropy',
    metrics=['accuracy']
)
```

input_shape= 28*28 = 784
first hidden layer with 512 neurons
784 * 512 = 401408
512 weights/parameters for bias
401408+512 = 401920
alt. (784+1) * 512 = 401920 # wynik dla dense1

```python
plot_model(
    model=network,
    # to_file='model.png', ->pozwala zapisać model do pliku
    show_shapes=True,
    show_dtype=False,
    show_layer_names=False,
    # rankdir='LR',
    expand_nested=False,
    dpi=100,  # dots per inch
    layer_range=None,
    show_layer_activations=True
)
```

```python
# redukcja wymiarowości tensora do macierzy
train_images = train_images.reshape((60000, 28 * 28))
train_images = train_images.astype('float32') / 255
print(train_images.shape)

test_images = test_images.reshape((10000, 28 * 28))
test_images = test_images.astype('float32') / 255
```

```python
from tensorflow.keras.utils import to_categorical  # get_dummies(), one-hot-encoder

train_labels = to_categorical(train_labels)
test_labels = to_categorical(test_labels)
```

```python
train_labels
```

```python
train_labels[0]
```

```python
train_labels.shape
```

```python
network.fit(
    train_images,
    train_labels,
    epochs=10,
    batch_size=128
)
```

```python
history = network.fit(
    train_images,
    train_labels,
    epochs=20,
    batch_size=313)
```

```python
history.history
```

```python
plt.plot(np.arange(1,21), history.history['loss'])
plt.title('loss changign by the epoch')
plt.ylabel('LOSS value')
plt.show()
```

```python
plt.plot(np.arange(1,21), history.history['accuracy'])
plt.title('ACC changign by the epoch')
plt.ylabel('ACC value')
plt.show()
```

```python
test_loss, test_acc = network.evaluate(test_images, test_labels)
print('test_acc:', test_acc)
```
