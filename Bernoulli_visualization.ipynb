```python
import numpy as np
from scipy.stats import bernoulli
import matplotlib.pyplot as plt
```

```python
plt.style.use('fivethirtyeight')
#data
probability = 0.83
size = 1_000
```

```python
fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(15, 4))

proba = bernoulli.rvs(probability, size=size) 

ax1.set_title("Random probability histogram")
ax1.hist(proba, weights=np.ones(len(proba)) / len(proba))

mean, var, skew, kurt = bernoulli.stats(probability, moments='mvsk')
print(f"Basic statistics: \nmean{mean}, \nvariable: {var},\
        \nskewness: {skew}, \nkurt: {kurt}")

arguments_x = np.arange(bernoulli.ppf(0., probability),
                bernoulli.ppf(1, probability)+1)


ax2.set_title('probability mass function')
ax2.scatter(arguments_x, bernoulli.pmf(arguments_x, probability),
            label='probability mass function', s=50.0)


x_arguments_points = np.linspace(-1, 2, 100)

ax3.set_title("distribution function")
ax3.plot(x_arguments_points, bernoulli.cdf(x_arguments_points, p=probability))

x_lim = [-1, 2]
ax1.set_xlim(x_lim)
ax2.set_xlim(x_lim)
ax3.set_xlim(x_lim)

y_lim = [0, 1.1]
ax1.set_ylim(y_lim)
ax2.set_ylim(y_lim)
ax3.set_ylim(y_lim)

plt.show()
```
